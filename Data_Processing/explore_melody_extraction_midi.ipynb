## Get the bar with maximum average pitch


### Shape of You



```python
fnm = eg_songs('Shape of You')
me = MelodyExtractor(fnm)

ic(me.fqs_ts, me.n_ts)
ic('seconds elapsed for a bar', me.n_ts / me.fqs_ts)
pr = me.pm.get_piano_roll(fs=me.fqs_ts)
ic(pr.shape)
Pmu.plot_piano_roll(me.pm, fqs=me.fqs_ts)

```
